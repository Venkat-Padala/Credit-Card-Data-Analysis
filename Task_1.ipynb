###Handling Outliers###

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,3))
sns.boxplot(x=df['MonthlyIncome'])
```

```python
plt.figure(figsize=(5,3))
sns.boxplot(x=df['age'])
```

```python
Q1_income = df['MonthlyIncome'].quantile(0.25)
Q3_income = df['MonthlyIncome'].quantile(0.75)
IQR_income = Q3_income - Q1_income
print(IQR_income)
lower_bound_income = Q1_income - 1.5 * IQR_income
upper_bound_income = Q3_income + 1.5 * IQR_income
print(lower_bound_income)
print(upper_bound_income)
```

```python
# Cap outliers
df['MonthlyIncome'] = np.where(df['MonthlyIncome'] > upper_bound_income, upper_bound_income, df['MonthlyIncome'])
df['MonthlyIncome'] = np.where(df['MonthlyIncome'] < lower_bound_income, lower_bound_income, df['MonthlyIncome'])
```

```python
lower_M=df['MonthlyIncome'].quantile(0.05)
df.loc[:,'MonthlyIncome']=df['MonthlyIncome'].clip(lower=lower_M)
```

```python
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
print(lower)
print(upper)
```

```python
upper_array = np.where(df['age'] >= upper)[0]
lower_array = np.where(df['age'] <= lower)[0]
```

```python
df.drop(index=upper_array, inplace=True)
df.drop(index=lower_array, inplace=True)
```
